# Deliverable 1: Preprocessing the Data for PCA

#### 1. Read in the crypto_data.csv to the Pandas DataFrame named crypto_df.

```python
# Load the crypto_data.csv dataset.
filepath = "../Cryptocurrencies/Cryptocurrencies/Resources/crypto_data.csv"
df= pd.read_csv(filepath)
df
```

```python
df.dtypes
```

#### 2. Keep all the cryptocurrencies that are being traded.

```python
# Keep all the cryptocurrencies that are being traded.
data_df = df[(df['IsTrading'] != False)]
data_df
```

```python
data_df['TotalCoinsMined'].notnull().count()
```

#### 3 Drop the IsTrading column.

```python
# Remove the "IsTrading" column. 
is_trading_df = data_df.drop(['IsTrading'], axis=1)
is_trading_df.set_index('Unnamed: 0',inplace=True)
is_trading_df
```

#### 4 Remove rows that have at least one null value.

```python
#remove Null Values 
null_df = is_trading_df.dropna()
null_df
#null_df.set_index('Unnamed: 0')
```

```python
for column in null_df.columns:
    print(f"Column {column} has {null_df[column].isnull().sum()} null values")
```

#### 5 Filter the crypto_df DataFrame so it only has rows where coins have been mined.

```python
null_df.columns
```

```python
# Keep the rows where coins are mined.
coined_mined_df = null_df[(null_df['TotalCoinsMined'] != 0)].dropna()
coined_mined_df
#coined_mined_df.set_index('Unnamed: 0')
```

```python
for column in coined_mined_df.columns:
    print(f"Column {column} has {coined_mined_df[column].isnull().sum()} null values")
```

#### 6. Create a new DataFrame that holds only the cryptocurrency names, and use the crypto_df DataFrame index as the index for this new DataFrame.

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
crypto_df = coined_mined_df['CoinName'] #NEED INDEX
crypto_df
```

```python
coin_names_df = pd.DataFrame(crypto_df)
coin_names_df.reset_index(drop=True, inplace=True)
#crypto_df.set_index('Unnamed: 0',inplace=True)
#crypto_df
coin_names_df
```

#### 7 Remove the CoinName column from the crypto_df DataFrame since it's not going to be used on the clustering algorithm.

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
# Remove the "IsTrading" column. 
coin_drop_df = coined_mined_df.drop(['CoinName'], axis=1, inplace=False )
coin_drop_df
#coin_drop_df.set_index('Unnamed: 0')
```

#### 8 Use the get_dummies() method to create variables for the two text features, Algorithm and ProofType, and store the resulting data in a new DataFrame named X.

```python
coin_drop_df.columns
```

```python
# Create our features
# Convert string columns to numbers and drop 'loan_status' column then assign to X
X = pd.get_dummies(coin_drop_df, columns=['Algorithm', 'ProofType'])#.drop('loan_status', axis=1)
X.head()
```

#### 9 Use the StandardScaler fit_transform() function to standardize the features from the X DataFrame.

```python
XS =StandardScaler().fit_transform(X)
print(XS[0:5])
```

```python
X.head(3)
```

# Deliverable 2: Reducing Data Dimensions Using PCA

```python
# Using PCA to reduce dimension to three principal components.
#Initialize PCA model
pca = PCA(n_components=3)
X_pca = pca.fit_transform(XS)
X_pca
```

```python
# Create a DataFrame with the three principal components.
PCA_df = pd.DataFrame(data=X_pca,columns=["PC1", "PC2", "PC3"],index=crypto_df)
PCA_df.head()
```

```python
# Initialize the K-Means model.
model = KMeans(n_clusters=4, random_state=0)
model
```

```python
# Create an elbow curve to find the best value for K.
#--Store Values of K to Plot
inertia = []
k = list(range(1, 11))

#--Loop Through K  Values and Find Inertia
# Looking for the best K
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(PCA_df)
    inertia.append(km.inertia_)
```
